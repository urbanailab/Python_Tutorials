# Step 3: Use Google API



```python
#If you run this code on your own PC, please run "pip install googlemaps" to get google map package first.
#I have installed all the packages in this google colab environment

import googlemaps
import requests
import io
from PIL import Image
from IPython.display import display
```

```python
#Here is the code sample of getting street view images
#Let's use a single coordinate first

# Replace with your own Google Maps API key
api_key = "ReplaceHereByYourKey"

#It's a coordinate in Tallahassee
latitude = 30.4383
longitude = -84.2807

#Change image size here
size = '1024x1024'

# Construct the URL for requesting the Street View image
street_view_url = f"https://maps.googleapis.com/maps/api/streetview?location={latitude},{longitude}&size={size}&key={api_key}"

# Send requests and open images
response_street_view = requests.get(street_view_url)
street_view_image = Image.open(io.BytesIO(response_street_view.content))

# Display the images
display(street_view_image)
```

```python
#If you have a list of coordinates

# List of coordinates (latitude, longitude)
coordinates = [
    (30.4383, -84.2807),  # Tallahassee
    (25.7617, -80.1918),  # Miami
    # Add more coordinates here
]

#Change image size here
size = '1024x1024'

# Loop through the coordinates and retrieve images
for index, (latitude, longitude) in enumerate(coordinates, start=1):
    # Construct the URL for requesting the Street View image
    street_view_url = f"https://maps.googleapis.com/maps/api/streetview?size={size}&location={latitude},{longitude}&key={api_key}"

    # Send requests and open images
    response_street_view = requests.get(street_view_url)
    street_view_image = Image.open(io.BytesIO(response_street_view.content))

    # Display the images
    print(f"Images for Coordinate {index}:")
    display(street_view_image)

    # Save images to a folder
    street_view_image.save(f"street_view_{index}.jpg")
    print(f"Images for Coordinate {index} saved.")

    print("-" * 50)
```

```python
#Here is the code sample of getting satellite images
#Let's also use a single coordinate first

#Replace with your own Google Maps API key
api_key = "ReplaceHereByYourKey"

#It's a coordinate in Tallahassee
latitude = 30.4385
longitude = -84.2809

#Change image size here
size = '1024x1024'

#Change the zoom level of image here
zoom_level = 15

# Construct the URL for requesting the satellite map image
satellite_map_url = f"https://maps.googleapis.com/maps/api/staticmap?center={latitude},{longitude}&zoom={zoom_level}&size={size}&maptype=satellite&key={api_key}"

# Send requests and open images
response_satellite_map = requests.get(satellite_map_url)
satellite_map_image = Image.open(io.BytesIO(response_satellite_map.content))

# Display the images
display(satellite_map_image)
```
